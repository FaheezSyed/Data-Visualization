```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/Hearing well-being Survey Report.csv")
df.head()
```

```python
# Check structure
df.info()

# Summary statistics
df.describe(include='all')

# Missing values
df.isnull().sum()
```

# Most columns are categorical .

there are missing values in Hearing_Test_Barrier

```python
# Find the mode of the 'Hearing_Test_Barrier' column
mode_value = df['Hearing_Test_Barrier'].mode()[0]

# Fill the missing values with the mode
df['Hearing_Test_Barrier'].fillna(mode_value, inplace=True)

# Verify that missing values have been filled
display(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Age_group", data=df)
plt.title("Distribution of Respondents by Age Group")
plt.show()
```

```python
df['Daily_Headphone_Use'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Daily Headphone Use")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Interest_in_Hearing_App", data=df)
plt.title("Interest in Hearing App")
plt.show()
```

```python
sns.histplot(df['Belief_Early_Hearing_Care'], bins=10, kde=True, color="skyblue")
plt.title("Belief in Early Hearing Care (Score Distribution)")
plt.show()
```

```python
sns.boxplot(x=df['Belief_Early_Hearing_Care'])
plt.title("Boxplot of Belief in Early Hearing Care")
plt.show()
```

```python
sns.kdeplot(df['Belief_Early_Hearing_Care'], fill=True, color="green")
plt.title("Density of Belief in Early Hearing Care")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Age_group", y="Belief_Early_Hearing_Care", data=df)
plt.title("Belief in Early Hearing Care by Age Group")
plt.show()
```

```python
sns.countplot(x="Daily_Headphone_Use", hue="Interest_in_Hearing_App", data=df)
plt.title("Interest in Hearing App vs Daily Headphone Use")
plt.show()
```
